Eu escolhi esse arquivo que é um recorte do IDH, pois quero evidenciar as desigualdades de gênero durante os anos de 2002 a 2021


```python
# Importando a biblioteca pandas
import pandas as pd
```

```python
# Abrir o arquivo excel
Base = '/content/idh.csv'
df = pd.read_csv(Base)
```

O Índice de Desenvolvimento Humano (IDH) compara indicadores de riqueza, alfabetização, educação, esperança de vida, natalidade e outros. Neste gráfico, podemos ver que o IDH masculino, traçado em azul, é sempre maior que o feminino, traçado em rosa. Podemos demonstrar o desigualdade de genero estrutural nos dados colhidos, já que o IDH masculino está sempre superior ao feminino. Podemos observar também que, desde 2017, o IDH geral vem diminuindo no Brasil.


```python
# Gráfico comparação da expectativa de vida feminina e masculina
plt.figure(figsize=(14, 7))
plt.plot(dfanos['ano_referencia'], dfanos['expectativa_de_vida_feminina'], marker='o', label='Expectativa de Vida Feminina', color='pink')
plt.plot(dfanos['ano_referencia'], dfanos['expectativa_de_vida_masculina'], marker='o', label='Expectativa de Vida Masculina', color='blue')
plt.title('Comparação da Expectativa de Vida Feminina e Masculina (2002-2021)')
plt.xlabel('Ano')
plt.ylabel('Expectativa de Vida')
plt.legend()
plt.grid(True)
plt.show()
```

Neste gráfico, podemos ver que a expectativa de vida das mulheres, traçada em rosa, é maior que a dos homens, traçada em azul. Também observamos que, desde o ano de 2017, a expectativa de vida tem vindo a reduzindo para ambos os gêneros e que por mais que a expectativa da vida da mulher seja maior o IDH permanece menor



```python
# Criar o gráfico da evolução da expectativa de anos de escola feminina e masculina com cores específicas
plt.figure(figsize=(14, 7))
plt.plot(dfanos['ano_referencia'], dfanos['expectativa_de_anos_escola_feminina'], marker='.', label='Expectativa de Anos de Escola Feminina', color='pink')
plt.plot(dfanos['ano_referencia'], dfanos['expectativa_de_anos_escola_masculina'], marker='.', label='Expectativa de Anos de Escola Masculina', color='blue')
plt.title('Comparação da Expectativa de Anos de Escola Feminina e Masculina (2002-2021)')
plt.xlabel('Ano')
plt.ylabel('Anos de Escola')
plt.legend()
plt.grid(True)
plt.show()
```

Neste gráfico, é evidente que as mulheres passam mais anos na escola do que os homens. Mesmo assim, ainda observamos muitas desigualdades de gênero, principalmente em relação aos salários entre homens e mulheres e mesmo as mulheres tendo mais expectativa de anos escolares os IDH continua menor.
